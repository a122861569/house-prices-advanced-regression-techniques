```python
import pandas as pd

train_data = pd.read_csv(r"C:\Users\a122861569\Desktop\kaggle\house-prices-advanced-regression-techniques\train.csv")
train_data.head()
```

```python
test_data = pd.read_csv(r"C:\Users\a122861569\Desktop\kaggle\house-prices-advanced-regression-techniques\test.csv")
test_data.head()
```

```python
print(train_data.columns)
```

```python
保存特质值test_ids,因为后面测试提交需要有id
保存目标变量
在训练集中删除目标列，并将测试集与训练集合并，方便统一化数据清洗
因为是合并的,所以根据len(y)可以实现数据集的分离，[:len(y)]及是训练集
```

```python
test_ids = test_data["Id"]
y = train_data["SalePrice"]
train_data.drop(["SalePrice"],axis=1,inplace = True)
combined = pd.concat([train_data,test_data],axis=0)
for col in combined.columns:
    if combined[col].dtype == "object":
        combined[col] = combined[col].fillna("None")
    else:
        combined[col] = combined[col].fillna(combined[col].median())
combined = pd.get_dummies(combined)

x_train = combined.iloc[:len(y),:]
x_test = combined.iloc[len(y) :,:]
```

```python
训练模型，填充训练集特征和目标变量
```

```python
from sklearn.ensemble import RandomForestRegressor
model = RandomForestRegressor(n_estimators=100,random_state=42)
model.fit(x_train,y)
```

```python
查看训练的精度
```

```python
from sklearn.model_selection import cross_val_score
scores = cross_val_score(model, x_train, y, cv=11)
print(f"Acurácia média (validação cruzada): {scores.mean():.3f} (+/- {scores.std() * 2:.3f})")
```

```python
predictions = model.predict(x_test)
submission = pd.DataFrame({
    "Id": test_ids,
    "SalePrice": predictions
})
submission.to_csv('submission1.csv', index=False)
print("Arquivo de submissão criado: submission.csv")
```
